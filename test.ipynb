```python
import cv2
import matplotlib.pyplot as plt
import keras
```

```python
image_cat = cv2.imread('dog.105.jpg')
plt.imshow(image_cat)
```

```python
image_resize = cv2.resize(image_cat,(150,150))
plt.imshow(image_resize)
```

```python
new_image_resize = image_resize.reshape((1,150,150,3))
new_image_resize.shape
```

```python
model_new = keras.models.load_model('classification_model.h5')
model_new.summary()
```

```python
import numpy as np
result = np.array(model_new.predict(new_image_resize))[0][0]
print(result)
```

```python
### 0 represents Dog & 1 represents Cat
if result>=0 and result<0.5:
    print('Dog')
else:
    print('Cat')
```
